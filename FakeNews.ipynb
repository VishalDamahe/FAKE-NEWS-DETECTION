**Feature Extraction**

focused on 'text' column

```python
import re    #regular expression
```

```python
def wordopt(text):
  # if text is None:
  #   return ""
  text = text.lower()   #convert text to lowercase

  #remove URLs
  text = re.sub(r'https?://\S+|www\.\S+', '', text)

  #remove HTML tags
  text = re.sub(r'<.*?>', '', text)

  #remove punctuations
  text = re.sub(r'[^\w\s]', '', text)

  #remove digits
  text = re.sub(r'\d', '', text)

  #remove newline characters
  text = re.sub(r'\n', ' ', text)

  return text
```

```python
news['text'] = news['text'].apply(wordopt)
```

```python
news['text']
```

```python
x = news['text']
y = news['label']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
x_train.shape
```

```python
x_test.shape
```

Text to Vectors of numerical data

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorization = TfidfVectorizer()
```

```python
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
xv_train
```

```python
xv_test
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression()
```

```python
LR.fit(xv_train, y_train)
```

```python
pred_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, pred_lr))
```

-Decision Tree CLassifier

```python
DTC.fit(xv_train, y_train)
```

```python
def output_lable(n):
  if n == 0:
    return "It is a Fake News"
  elif n == 1:
    return "It is a Genuine News"
```

```python
def manual_testing(news):
  testing_news = {"text":[news]}  #create a dictionary
  new_def_test = pd.DataFrame(testing_news)
  new_def_test["text"] = new_def_test["text"].apply(wordopt)
  new_x_test = new_def_test["text"]
  new_xv_test = vectorization.transform(new_x_test)   #lets assume 'vectorization' is the vectorizer object
  pred_lr = LR.predict(new_xv_test)
  pred_gbc = gbc.predict(new_xv_test)
  pred_rfc = rfc.predict(new_xv_test)
  return "\n\nLR Prediction: {} \nGBC Prediction: {} \nRFC Prediction: {}".format(output_lable(pred_lr[0]), output_lable(pred_gbc[0]), output_lable(pred_rfc[0]))
```
